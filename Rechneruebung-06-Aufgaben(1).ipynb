Sie enthält auch zwei Funktionen, `Umfang` und `Flaeche`. Der Umfang eines Polgons ist einfach die Summe aller Abstände aller Punkte zu ihrem Nachbarpunkt (immer in ein und dieselbe Richtung). Für die Fläche beliebiger Polygone gibt es keine einfache Formel, daher ist die Funktion erst mal "leer", wird aber von den Unterklassen überschrieben.

Die Funktion `Umfang` sollen Sie selbt programmieren: Wenn mit $d(p_i,p_j)$ der Abstand der Punkte $p_i$ und $p_j$ gemeint ist, und das Polygon die Eckpunkte $p_0, p_1, \dots p_{n-1}$ hat, dann ist der Umfang einfach: $U = d(p_0,p_{n-1})+\sum\limits_{i=0}^{n-2} d(p_i,p_{i+1})$.
