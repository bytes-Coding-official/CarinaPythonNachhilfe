## Aufgabe 2: Rekursive Punktrechnung

### a) Multiplikation

Die Multiplikation zweier natürlicher Zahlen $m$ und $n$ ist wie folgt definiert:

$m\!\cdot\! n\! = \left\{ \begin{array}{rl} 0 & \text{ falls }n\!=\!0 \\ m\!+\!m\!\cdot\!(n\!-\!1) & \text{ sonst}\end{array}\right.$

Man kann erkennen, dass das Produkt von $m$ und $n$ Null ist für den Fall, dass $n=0$ ist. Und sonst ist es $m$ mal mehr als das Produkt von $m$ mit $n-1$. So ist zum Beispiel $7\cdot 5 = 7+7\cdot(5-1)$. Das ist ihnen bestimmt schon seit der 2. Klasse geläufig.

Programmieren Sie nun eine Funktion `produkt(m,n)`, die gemäß der obigen Definition das Produkt ihrer beiden Argumente berechnet und zurückgibt. Selbstverständlich dürfen Sie **kein Multiplikationszeichen** im Funktionsrumpf verwenden.

````python
def Produkt(m,n):
    if n == 0:
        return 0
    else:
        return m+Produkt(m, n-1)
    
print(Produkt(2,3))            # Erwartung: Es wird 6 ausgegeben
print(Produkt(5,Produkt(4,2))) # Erwartung: Es wird 40 ausgegeben
print(Produkt(1,7))            # Erwartung: Es wird 7 ausgegeben
print(Produkt(0,4))            # Erwartung: Es wird 0 ausgegeben
````

### b) Division

Wenn Sie die Multiplikation hingekriegt haben, dann können Sie bestimmt auch die rekursive Definition der *Ganzzahl*division in ein Python-Programm umsetzen:

$m/n = \left\{ \begin{array}{rl} 0 & \text{ falls }n>m \\ 1+(m-n)/n & \text{ sonst}\end{array}\right.$

Vom Hinschauen sieht man: Wenn man $m/n$ dividiert kommt entweder 0 heraus (nämlich, wenn $n>m$) oder es kommt 1 mehr heraus, als wenn man $m-n$ durch $n$ dividieren würde. So ist zum Beispiel $13/4$ eins mehr als $(13-4)/4$ - wenn man 13 Äpfel an 4 Kinder verteilt, kriegt jedes einen Apfel mehr, als wenn man $(13-4)$ Äpfel verteilen würde.

Programmieren Sie die rekursive Funktion `Quotient(m,n)` und verwenden dabei selbstverständlich kein Divisionssymbol im Funktionsrumpf. 

Bemerkung: Natürlich dürfen Sie sich auf die Schulter klopfen, wenn Sie es irgendwie trickreich hinkriegen, die Ganzzahldivision ohne Divisionsoperator zu programmieren. Aber wenn sie nicht rekursiv ist, dann ist die Aufgabe nicht gelöst.

````python
def Quotient(m,n):
    # Hier kommen Sie
    if n > m: 
        return 0
    else:
        return 1+Quotient(m-n, n)
  
    
print(Quotient(13,4))              # Erwartung: Es wird 3 ausgegeben
print(Quotient(12,Quotient(15,5))) # Erwartung: Es wird 4 ausgegeben
print(Quotient(4,7))               # Erwartung: Es wird 0 ausgegeben
print(Quotient(8,1))               # Erwartung: Es wird 8 ausgegeben
````

## Aufgabe 3: Das Pascalsche Dreieck

Unten ist das Pascalsche Dreieck dargestellt (benannt nach Blaise Pascal). Um es zu schreiben, beginnt man mit einer 1, schreibt schräg nach links und rechts versetzt drunter zwei weitere Einsen. Dann fügt man immer weiter neue Zeilen hinzu, die immer eine Zahl mehr enthalten als die direkt darüber liegende Zeile. Am Anfang und Ende jeder Zeile steht immer eine 1. Dazwischen steht immer die Summe der beiden schräg darüber stehenden Zahlen.

Man bezeichnet die in der $n$-ten Zeile an $k$-ter Stelle stehende Zahl auch als "$n$ über $k$", geschrieben. 
$\left(\begin{array}{c}n\\k\end{array}\right)$. 

Im Beispiel nebenan ist die fünfte Zeile (Zählen beginnt bei 0) rosa unterlegt, und die jeweils zweiten Werte jeder Zeile (= "die zweite Spalte") sind gelb unterlegt. Der Wert $\left(\begin{array}{c}5\\2\end{array}\right)$ ist dort als 10 abzulesen.

Programmieren Sie die Funktion `p(n,k)`, indem Sie im untenstehenden Skelett die ausgeixten Teile passend ersetzen, so dass der Funktionswert $\left(\begin{array}{c}n\\k\end{array}\right)$ berechnet wird. Die Funktion muss rekursiv arbeiten - falls Sie im Internet etwas mit Fakultät und so finden, ist das toll, aber gilt jetzt nicht als Lösung.

````python
def p(n,k):
    if k ==0 or n==k:
        return 1
    else:
        return p(n-1, k-1) + p(n-1, k)

print(p(5,2))
````
